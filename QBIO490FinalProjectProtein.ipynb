```python
import cptac
import numpy as np
import pandas as pd
cptac.list_datasets()
cptac.download(dataset="Luad")
luad = cptac.Luad()
luad.list_data()
```

```python
protein_data = luad.get_proteomics() # get the proteomics data
protein_data
protein_data.columns = protein_data.columns.get_level_values(0) 
protein_data
```

```python
protein_data = pd.DataFrame(protein_data)
protein_data = protein_data.dropna(axis="columns")
protein_data
```

```python
rna_data = luad.get_transcriptomics()
rna_data
```

```python
clinical_data = luad.get_clinical()
clinical_data
```

```python
Stage_I_mask = np.where(clinical_data.loc[:,"Stage"] == "1A", True,\
                        np.where(clinical_data.loc[:,"Stage"] == "1B",True,False))
clinical_data.loc[Stage_I_mask,"Stage"] = "Stage I"
Stage_II_mask = np.where(clinical_data.loc[:,"Stage"] == "2A", True,\
                        np.where(clinical_data.loc[:,"Stage"] == "2B",True,False))
clinical_data.loc[Stage_II_mask,"Stage"] = "Stage II"
Stage_III_mask = np.where(clinical_data.loc[:,"Stage"] == "3A", True,\
                        np.where(clinical_data.loc[:,"Stage"] == "3B",True,False))
clinical_data.loc[Stage_III_mask,"Stage"] = "Stage III"
Stage_IV_mask = np.where(clinical_data.loc[:,"Stage"] == "4", True, False)
clinical_data.loc[Stage_IV_mask,"Stage"] = "Stage IV"
```

```python
clinical_data = pd.DataFrame(clinical_data)
clinical_data = clinical_data[clinical_data['Stage'].notna()]
clinical_data
```

```python
Stage_I_data = clinical_data.loc[clinical_data["Stage"] == "Stage I"]
Stage_I_patients = Stage_I_data.index
Stage_II_data = clinical_data.loc[clinical_data["Stage"] == "Stage II"]
Stage_II_patients = Stage_II_data.index
Stage_III_data = clinical_data.loc[clinical_data["Stage"] == "Stage III"]
Stage_III_patients = Stage_III_data.index
Stage_IV_data = clinical_data.loc[clinical_data["Stage"] == "Stage IV"]
Stage_IV_patients = Stage_IV_data.index
```

```python
Stage_I_protein = protein_data.loc[Stage_I_patients]
Stage_II_protein = protein_data.loc[Stage_II_patients]
Stage_III_protein = protein_data.loc[Stage_III_patients]
Stage_IV_protein = protein_data.loc[Stage_IV_patients]
```

```python
Stage_II_protein
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Stage_I_int = scaler.fit_transform(Stage_I_protein)
Stage_II_int = scaler.fit_transform(Stage_II_protein)
Stage_III_int = scaler.fit_transform(Stage_III_protein)
Stage_I_int = pd.DataFrame(Stage_I_int)
Stage_II_int = pd.DataFrame(Stage_II_int)
Stage_III_int = pd.DataFrame(Stage_III_int)

Stage_I_int.index = Stage_I_protein.index
Stage_II_int.index = Stage_II_protein.index
Stage_III_int.index = Stage_III_protein.index

Stage_I_int.columns = Stage_I_protein.columns
Stage_II_int.columns = Stage_II_protein.columns
Stage_III_int.columns = Stage_III_protein.columns

Stage_I_protein = Stage_I_int
Stage_II_protein = Stage_II_int
Stage_III_protein = Stage_III_int
Stage_I_protein
```

```python
import scipy.stats as stats
x = stats.f_oneway(Stage_I_protein.loc[:, Stage_I_protein.columns[1]], Stage_II_protein.loc[:, Stage_II_protein.columns[1]], Stage_III_protein.loc[:,Stage_III_protein.columns[1]])
x
```

```python
protein_gene = []
protein_pvals = []
for i in range(len(protein_data.columns)):
    x = stats.f_oneway(Stage_I_protein.loc[:, Stage_I_protein.columns[i]], Stage_II_protein.loc[:,Stage_II_protein.columns[i] ], Stage_III_protein.loc[:, Stage_III_protein.columns[i]])
    x = x.pvalue
    if (isinstance(x, np.float64) == False):
        for j in range(len(x)):
            if (x[j] < 1):
                protein_gene.append(i)
                protein_pvals.append(x[j])
                break
    else:
        if (x < 1):
            protein_gene.append(i)
            protein_pvals.append(x)
                
protein_gene
```

```python
Stage_I_rna = rna_data.loc[Stage_I_protein.index,:]
Stage_II_rna = rna_data.loc[Stage_II_protein.index,:]
Stage_III_rna = rna_data.loc[Stage_III_protein.index,:]
```

```python
rna_names_CPTAC= np.intersect1d(rna_data.columns, rna_names)
rna_names_common = np.intersect1d(Stage_I_protein.columns, rna_names_CPTAC)
mutation_gene = ["MUC16","CSMD3","XIRP2","TTN","ZFHX4"]
mutation_names_CPTAC = np.intersect1d(rna_data.columns, mutation_gene)
mutation_names_common = np.intersect1d(Stage_I_protein.columns, mutation_names_CPTAC)
mutation_names_common
```

```python
sig_gene_common = ['C9', 'CRP', 'KRT1', 'PZP','MUC16', 'XIRP2','TTN']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_df = pd.DataFrame(np.ndarray(shape=(7, 7), dtype=np.float64),
                      index = sig_gene_common,
                      columns = sig_gene_common)

# 2. fill in the data frame!
for p1 in sig_gene_common:
    for p2 in sig_gene_common:
        # calculate the correlations between protein and RNA
        # then, use .loc[] to store the correlation in corr_df
        corr,pval = stats.spearmanr(Stage_I_rna[p1],Stage_I_protein[p2],nan_policy="omit")
        corr_df.loc[p1,p2] =corr

# 3. create the heat map
plot = sns.heatmap(
    corr_df,
    cmap='mako',
)
plot.set_xlabel('Protein', fontsize=12)
plot.set_ylabel('gene', fontsize=12)
#addd title
plot.set_title('StageI gene vs Protein expression', fontsize=17)
plt.show()
plt.savefig('StageI gene vs Protein expression.png')
```
